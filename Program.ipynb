```python
import pandas as pd
from datetime import datetime
import os 
```

```python
os.makedirs("data",exist_ok=True)
d=datetime.now()
```

```python
ACCOUNT_PATH="data/accounts.csv"
TRANSACTIONS_PATH="data/transactions.csv"
BALANCES_PATH="data/daily_balances.csv"
USERS_PATH="data/users.csv"
```

```python
accounts=pd.read_csv(ACCOUNT_PATH)
transactions=pd.read_csv(TRANSACTIONS_PATH)
daily_balancers=pd.read_csv(BALANCES_PATH)
users=pd.read_csv(USERS_PATH)
```

```python
user_session_id=1
```

```python
def add_user( name,user_type,password=""):
    global users

    next_id=1 if users.empty else  users["users_id"].max()+1
    new_row={
        "user_id":next_id,
        "name":name,
        "type":user_type,
        "password":password
    }

    users=pd.concat([users,pd.DataFrame([new_row])], ignore_index=True)
    users.to_csv(USERS_PATH, index=False)
    print("add user successfully")
```

```python
add_user("Myo Aung","Personal","PaPa")
```

```python
def add_account(account_name,account_type,group,user_id=""):
    global accounts

    next_id=1 if accounts.empty else  accounts["account_id"].max()+1
    new_row={
        "account_id":next_id,
        "account_name":account_name,
        "account_type":account_type,
        "group":group,
        "user_id":user_id
    }

    accounts=pd.concat([accounts,pd.DataFrame([new_row])], ignore_index=True)
    accounts.to_csv(ACCOUNT_PATH,index=False)
    print("add account successfully")
```

```python
add_account("Montery","Credit","lone",user_session_id)
```

```python
def add_transaction(t_type,amount,account_id,category,note,user_id=""):
    global transactions
    global d
    next_id=1 if accounts.empty else  accounts["account_id"].max()+1
    new_row={
        "txn_id":next_id,
        "date":d,
        "type":t_type,
        "account_id":account_id,
        "amount":amount,
        "category":category,
        "note":note,
        "user_id":user_id
    }

    transactions=pd.concat([transactions,pd.DataFrame([new_row])], ignore_index=True)
    transactions.to_csv(ACCOUNT_PATH,index=False)
    print("add transaction successfully")
```
