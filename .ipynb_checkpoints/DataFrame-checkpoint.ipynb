```python
import pandas as pd
```

```python
transactions = pd.DataFrame(columns=[
    "txn_id","date","type","amount","account_id","category","note"
])
```

```python

accounts = pd.DataFrame(columns=[
    "account_id",
    "account_name",   # "Chase Checking", "Affirm Loan", "Car Loan"
    "account_type",   # "asset" or "debt"
    "group"           # optional: "cash", "bank", "credit", "loan", "investment"
])
 

```

```python
daily_balances = pd.DataFrame(columns=[
    "id",
    "date",
    "account_id",
    "balance"  # assets: amount you own, debts: amount you owe
])
```

```python
transactions.to_csv("./data/transactions.csv",index=False)
```

```python
accounts.to_csv("./data/accounts.csv",index=False)
```

```python
daily_balances.to_csv("./data/daily_balances.csv",index=False)
```
